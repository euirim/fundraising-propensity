```python
%matplotlib inline
import os

import matplotlib.pyplot as plt

from main import create_hists_for_dataset
```

```python
alpha = 0.4
fig = plt.figure(edgecolor='k', figsize=(6,5), dpi=300)
create_hists_for_dataset('word2vec', 'word2vec', fig, alpha, num_rows=4, num_cols=1)
create_hists_for_dataset('doc2vec', 'doc2vec', fig, alpha, num_rows=4, num_cols=1)
create_hists_for_dataset('bert', 'BERT', fig, alpha, num_rows=4, num_cols=1)
fig.subplots_adjust(wspace=0.5)
fig.subplots_adjust(hspace=1.0)
fig.savefig('./tmp/error_hist_across_datasets_overlay.png', bbox_inches='tight', dpi=500)
```
